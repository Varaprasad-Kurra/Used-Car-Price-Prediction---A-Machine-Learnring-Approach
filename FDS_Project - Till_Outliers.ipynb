```python
#Obtaining the Data Quality report
#For Data Data Quality report the attributes in the table various for Contionous and Categorical Data.

#For Continous Data = {Count, %Miss, Card, Min, 1st_Qrt, Mean, Median, 3rd_Qrt, Max, Std.Dev} are the attributes in Data Quality Report
#For Categorical Data = {Count, %Miss, Card, Mode, Mode_Freq, Mode_%, 2nd_Mode, 2nd_Mode_Freq, 2nd_Mode_%} are the attributes in Data Quality Report

# Let us look into the Continous Data Data Quality Report

def data_quality_Continous(df):
    df_report = pd.DataFrame()
    
    df_report['Count'] = df.count()                              
    df_report['%Miss'] = round((df.isnull().sum()/ df.count()) * (100),3)  
    df_report['Card.'] = df.apply(pd.Series.nunique)             
    df_report['Min'] = df.min()                                  
    df_report['1st_Qrt'] = df.quantile([0.25]).T                 
    df_report['Mean'] = round(df.mean(),2)                                
    df_report['Median'] = df.median()                            
    df_report['3rd_Qrt'] = df.quantile([0.75]).T                 
    df_report['Max'] = df.max()                                  
    df_report['Std.Dev.'] = round(df.std(),2)                             
    
    return df_report

df_cont = data[['Year','Kilometers_Driven','Mileage','Engine','Power','New_Price','Price']]
df_cont_report = data_quality_Continous(df_cont)
display(df_cont_report)
```

```python
def get_cate_data_quality_report(cat_data):
    df_report = pd.DataFrame()

    df_report['Count'] = cat_data.count()                   
    df_report['%Miss'] = round(pd.isnull(cat_data).sum() * 100 / cat_data.count(),3)
    df_report['Card.'] = cat_data.apply(pd.Series.nunique)  

    cnt = 0
    cat_attr_dict = {}
    for (feature, series) in cat_data.iteritems():
        temp_dict = {}
        for i in series.tolist():
            if i not in temp_dict:
                temp_dict[i] = 0
            temp_dict[i] += 1
        cat_attr_dict[feature] = sorted(temp_dict.items(), key=lambda x: x[1], reverse=True)        
        cnt += 1

    attr_mode_dict = {}
    for name in cat_attr_dict:
        temp_arr = []
        temp_arr.append(cat_attr_dict[name][0][0])    
        temp_arr.append(cat_attr_dict[name][0][1])
        temp_arr.append(cat_attr_dict[name][1][0])
        temp_arr.append(cat_attr_dict[name][1][1])
        attr_mode_dict[name] = temp_arr

    df_mode = pd.DataFrame(attr_mode_dict).T
    df_report['Mode'] = df_mode.iloc[:, 0] 
    df_report['Mode_Freq.'] = df_mode.iloc[:, 1]
    df_report['Mode_%'] = df_report['Mode_Freq.'] * 100 / cat_data.count()
    df_report['2nd_Mode'] = df_mode.iloc[:, 2]
    df_report['2nd_Mode_Freq.'] = df_mode.iloc[:, 3]
    df_report['2nd_Mode_%'] = df_report['2nd_Mode_Freq.'] * 100 / cat_data.count()
    return df_report

cate_features = ['Location','Fuel_Type','Transmission','Owner_Type','Seats']
df_cate = data[cate_features]
print("\n[ Cate Data Quality Report of Second Hand Cars dataset ]\n")
df_cate_report = get_cate_data_quality_report(df_cate)
display(df_cate_report)
```

# Preprocessing the Data

```python
# Name of the CAR, We have observed that the name of Car is too long and there are observation of count more than 4,000. 
# We can crop the name of the car to it's Brand name.
new_data = data.copy()
new_data['Name']=new_data['Name'].apply(lambda Name: str(Name).split(" ")[0])

plt.figure(figsize=(10,8))     
plt.subplot(3,1,1)
new_data['Name'].value_counts()
new_data['Name'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Seats')
plt.xticks(rotation=90)
plt.ylabel('Count of Seats in Cars')
plt.title('Bar plot for Car distribution')

new_data['Mileage']=new_data['Mileage'].apply(lambda Mileage: str(Mileage).split(" ")[0]) #Splits values seperated by ' ' and Replaces Null values to Zero.
new_data['Engine']=new_data['Engine'].apply(lambda Engine: str(Engine).split(" ")[0])     #Splits values seperated by ' ' and Replaces Null values to Zero.
new_data['Power']=new_data['Power'].apply(lambda Power: str(Power).split(" ")[0])         #Splits values seperated by ' ' and Replaces Null values to Zero.
new_data.rename(columns = {'Engine':'Engine_CC', 'Power':'Power_bhp'},inplace = True) 
new_data.head()


#We have succesfully renamed the Columns to make the better understanding 
#We have set the carname based in the Brand name
```

# Handling the Missing Values

```python
#We will replace seats and New_price with Mean Values
new_data['Seats'].fillna(new_data['Seats'].mean().astype(int), inplace=True)

for i in new_data.columns:
    missing_values = new_data[i].isnull().sum()
    print('Missing Values of ',i , 'is     ',missing_values)
print('')

new_data.drop(['New_Price'], axis=1, inplace = True)
# We dropped the column 

#But we have have replaced Mileage	Engine_CC	Power_bhp na values with 0. We need to impute them with either mean/Median/ We chose mean here.
# We need to cconvert the type of Mileage	Engine_CC	Power_bhp. It was given they are object. But they are contious data.


new_data['Engine_CC'] = new_data.Engine_CC.astype(float)

new_data['Mileage'] = new_data.Mileage.astype(float)

new_data['Power_bhp'] = new_data['Power_bhp'].replace('null', 0)

new_data['Power_bhp'] = new_data.Power_bhp.astype(float)
new_data['Power_bhp'].fillna(new_data['Power_bhp'].mean().astype(float), inplace=True)
new_data['Mileage'].fillna(new_data['Mileage'].mean().astype(float), inplace=True)
new_data['Engine_CC'].fillna(new_data['Engine_CC'].mean().astype(float), inplace=True)


new_data.info()
```

```python
new_cont_report['IQR'] = new_cont_report['3rd_Qrt'] - new_cont_report['1st_Qrt']
new_cont_report['outlier_lower_bound'] = round(new_cont_report['1st_Qrt'] - 1.5*new_cont_report['IQR'],2)
new_cont_report['outlier_upper_bound'] = round(new_cont_report['3rd_Qrt'] + 1.5*new_cont_report['IQR'],2)
display(new_cont_report)

df_cont_clamped = pd.DataFrame()
for (name, series) in new_cont.iteritems():
    oub = new_cont_report.loc[name, 'outlier_upper_bound']
    olb = new_cont_report.loc[name, 'outlier_lower_bound']
    df_cont_clamped[name] = series.clip(oub, olb)
display(df_cont_clamped)
```
